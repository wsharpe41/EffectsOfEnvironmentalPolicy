```python
import pandas as pd
import os
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import warnings
```

```python
# import all data that need to be processed
# Air and Climate
files_air_and_climate = os.listdir('../data/Air and Climate')
df_name_air_and_climate = []
# import csv with '_preprocessed' in name but "_imputed" not in name as pd.DataFrame
for file in files_air_and_climate:
    if '_processed' in file and '_imputed' not in file:
        globals()[file[:-4]] = pd.read_csv('../data/Air and Climate/' + file)
        df_name_air_and_climate.append(file[:-4])
        # print dataframe name
        print(file[:-4])
# Inland Water Resources
files_inland_water_resources = os.listdir('../data/Inland Water Resources')
df_name_inland_water_resources = []
# import csv with '_preprocessed' in name as pd.DataFrame
for file in files_inland_water_resources:
    if '_processed' in file and '_imputed' not in file:
        globals()[file[:-4]] = pd.read_csv('../data/Inland Water Resources/' + file)
        df_name_inland_water_resources.append(file[:-4])
        # print dataframe name
        print(file[:-4])
# Waste
files_waste = os.listdir('../data/Waste')
df_name_waste = []
# import csv with '_preprocessed' in name as pd.DataFrame
for file in files_waste:
    if '_processed' in file and '_imputed' not in file:
        globals()[file[:-4]] = pd.read_csv('../data/Waste/' + file)
        df_name_waste.append(file[:-4])
        # print dataframe name
        print(file[:-4])
```

```python
# convert '...' and 'NaN' into np.nan
# remove ',' in numbers
for df_name in df_name_air_and_climate:
    globals()[df_name] = globals()[df_name].replace('...', np.nan)
    globals()[df_name] = globals()[df_name].replace('NaN', np.nan)
    globals()[df_name] = globals()[df_name].replace(',', '', regex=True)
for df_name in df_name_inland_water_resources:
    globals()[df_name] = globals()[df_name].replace('...', np.nan)
    globals()[df_name] = globals()[df_name].replace('NaN', np.nan)
    globals()[df_name] = globals()[df_name].replace(',', '', regex=True)
for df_name in df_name_waste:
    globals()[df_name] = globals()[df_name].replace('...', np.nan)
    globals()[df_name] = globals()[df_name].replace('NaN', np.nan)
    globals()[df_name] = globals()[df_name].replace(',', '', regex=True)
```

```python
# Impute missing values
# Initiate warning control
warnings.filterwarnings('error')
# Air and Climate
for df_name in df_name_air_and_climate:
    ncol = 31
    ncol_percentage = 31 -2 -1
    for row in range(globals()[df_name].shape[0]):
        # if the row has less than 3 non-NaN values or all values are non-NaN, skip
        if globals()[df_name].iloc[row, 2:ncol].notnull().sum() < 3:
            continue
        elif globals()[df_name].iloc[row, 2:ncol].isnull().sum() == 0:
            try: 
                # read the final 28 column names and fill in the percentage change according to the first part of the column name
                col_name = globals()[df_name].columns
                for i in range(1, ncol_percentage+1):
                    per_year_name = col_name[i*-1]
                    # access the first part of it by splitting the string
                    year_name = int(per_year_name.split(' ')[0])
                    # calculate the percentage change
                    globals()[df_name].at[row, per_year_name] = (globals()[df_name].at[row, str(year_name+1)] - globals()[df_name].at[row, str(year_name)]) / globals()[df_name].at[row, str(year_name)]
            except RuntimeWarning:
                print('RuntimeWarning: probably due to division by zero in row ' + str(row) + ' of ' + df_name)
        else:
            # fit an logistic regression model to predict the missing values
            # select data that is not NaN in the row as training data
            X_train = globals()[df_name].iloc[row, 2:ncol][~globals()[df_name].iloc[row, 2:ncol].isnull()].index.values.astype('int64').reshape(-1, 1)
            y_train = globals()[df_name].iloc[row, 2:ncol][~globals()[df_name].iloc[row, 2:ncol].isnull()].values.astype('float64')
            # fit a Polynomial Regression model
            poly_reg = PolynomialFeatures(degree=2)
            X_poly = poly_reg.fit_transform(X_train)
            lin_reg = LinearRegression()
            lin_reg.fit(X_poly, y_train)
            # predict the missing values
            X_test = globals()[df_name].iloc[row, 2:ncol][globals()[df_name].iloc[row, 2:ncol].isnull()].index.values.astype('int64').reshape(-1, 1)
            y_pred = lin_reg.predict(poly_reg.fit_transform(X_test))
            try: 
                # read the final 28 column names and fill in the percentage change according to the first part of the column name
                col_name = globals()[df_name].columns
                for i in range(1, ncol_percentage+1):
                    per_year_name = col_name[i*-1]
                    # access the first part of it by splitting the string
                    year_name = int(per_year_name.split(' ')[0])
                    # calculate the percentage change
                    globals()[df_name].at[row, per_year_name] = (globals()[df_name].at[row, str(year_name+1)] - globals()[df_name].at[row, str(year_name)]) / globals()[df_name].at[row, str(year_name)]
            except RuntimeWarning:
                print('RuntimeWarning: probably due to division by zero in row ' + str(row) + ' of ' + df_name)
    # save the imputed dataframe
    globals()[df_name].to_csv('../data/Air and Climate/' + df_name + '_imputed.csv', index=False)
    print(df_name + ' imputed')

            
            



        
        
```

```python
# Inland Water Resources
for df_name in df_name_inland_water_resources:
    ncol = 26
    ncol_percentage = 17
    for row in range(globals()[df_name].shape[0]):
        # if the row has less than 3 non-NaN values or all values are non-NaN, skip
        if globals()[df_name].iloc[row, 2:26].notnull().sum() < 3:
            # print(df_name + ' row ' + str(row) + ' skipped')
            continue
        elif globals()[df_name].iloc[row, 2:26].isnull().sum() == 0:
            try: 
                # read the final 28 column names and fill in the percentage change according to the first part of the column name
                col_name = globals()[df_name].columns
                for i in range(1, ncol_percentage+1):
                    per_year_name = col_name[i*-1]
                    # access the first part of it by splitting the string
                    year_name = int(per_year_name.split(' ')[0])
                    # calculate the percentage change
                    globals()[df_name].at[row, per_year_name] = (globals()[df_name].at[row, str(year_name+1)] - globals()[df_name].at[row, str(year_name)]) / globals()[df_name].at[row, str(year_name)]
            except RuntimeWarning:
                print('RuntimeWarning: probably due to division by zero in row ' + str(row) + ' of ' + df_name)
        else:
            # fit an logistic regression model to predict the missing values
            # select data that is not NaN in the row as training data
            X_train = globals()[df_name].iloc[row, 2:ncol][~globals()[df_name].iloc[row, 2:ncol].isnull()].index.values.astype('int64').reshape(-1, 1)
            y_train = globals()[df_name].iloc[row, 2:ncol][~globals()[df_name].iloc[row, 2:ncol].isnull()].values.astype('float64')
            # fit a Polynomial Regression model
            poly_reg = PolynomialFeatures(degree=2)
            X_poly = poly_reg.fit_transform(X_train)
            lin_reg = LinearRegression()
            lin_reg.fit(X_poly, y_train)
            # predict the missing values
            X_test = globals()[df_name].iloc[row, 2:ncol][globals()[df_name].iloc[row, 2:ncol].isnull()].index.values.astype('int64').reshape(-1, 1)
            y_pred = lin_reg.predict(poly_reg.fit_transform(X_test))
            # update the original dataframe based on X_test and y_pred using DataFrame.at
            for i in range(len(X_test)):
                globals()[df_name].at[row, str(X_test[i][0])] = y_pred[i]
            try: 
                # read the final 28 column names and fill in the percentage change according to the first part of the column name
                col_name = globals()[df_name].columns
                for i in range(1, ncol_percentage+1):
                    per_year_name = col_name[i*-1]
                    # access the first part of it by splitting the string
                    year_name = int(per_year_name.split(' ')[0])
                    # calculate the percentage change
                    globals()[df_name].at[row, per_year_name] = (globals()[df_name].at[row, str(year_name+1)] - globals()[df_name].at[row, str(year_name)]) / globals()[df_name].at[row, str(year_name)]
            except RuntimeWarning:
                print('RuntimeWarning: probably due to division by zero in row ' + str(row) + ' of ' + df_name)
    # save the imputed dataframe
    globals()[df_name].to_csv('../data/Inland Water Resources/' + df_name + '_imputed.csv', index=False)
    print(df_name + ' imputed')
```
